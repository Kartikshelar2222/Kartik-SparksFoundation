#  Checking Co-realtion in our Data

```python
df.corr()
sns.heatmap(df.corr())
```

```python
# Dropping the country and Postal code out of the dataset ,Since they are not affecting our profit
df = df.drop(['Country','Postal Code'],axis=1)
```

```python
# Changing the datatype of Qunatity
df['Quantity'] = df['Quantity'].astype('float')
```

```python
# Scatter Plot between Sales and Profit
sns.scatterplot(data=df,x='Sales',y='Profit')
```

```python
# Scatter Plot between Discount and Profit
sns.scatterplot(data=df,x='Discount',y='Profit')
```

```python
sns.lineplot(data=df,x='Sales',y='Profit')
```

```python
sns.lineplot(data=df,x='Discount',y='Profit')
```
